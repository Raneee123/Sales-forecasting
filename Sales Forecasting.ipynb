```python
y=newdf['Item_Outlet_Sales']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
def train(model,x,y):
    #training the model
    model.fit(x,y)
    predict=model.predict(x)
    # perform cross-validation
    cv_score= cross_val_score(model,x,y,scoring='neg_mean_squared_error',cv=5)
    print('Model Report')
    print('scoring- neg_mean_squared_error')
    print(cv_score)
    cv_score=np.abs(np.mean(cv_score))
    print('ABS Average of -neg_mean_squared_error', cv_score)
    cv_score=cross_val_score(model,x,y,cv=5)
    print()
    print('R2 score')
    print(cv_score)
    cv_score=np.mean(cv_score)
    print('Average R2 score', cv_score)
    print()
    print('Accuracy for full data')
    print('R2_score', r2_score(y,predict))
    print()
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
```

```python
model=LinearRegression(normalize=True)
train(model,x_train,y_train)
```

```python
coeff=pd.Series(model.coef_,x.columns).sort_values()
coeff.plot(kind='bar',title='Model Coefficient')
plt.show()
```

```python
model=Ridge(normalize=True)
train(model,x_train,y_train)
coeff=pd.Series(model.coef_,x.columns).sort_values()
coeff.plot(kind='bar',title='Model Coefficient')
plt.show()
    
```

```python
model=Lasso()
train(model,x_train,y_train)
coeff=pd.Series(model.coef_,x.columns).sort_values()
coeff.plot(kind='bar',title='Model Coefficient')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
train(model,x_train,y_train)
coeff=pd.Series(model.feature_importances_,x.columns).sort_values(ascending=False)
coeff.plot(kind='bar',title='Feature Importances')
plt.show()
```

```python
from sklearn.model_selection import RandomizedSearchCV
max_features=['auto','sqrt']
max_depth=[int(x) for x in np.linspace(5,30, num=6)]
min_samples_split=[2,5,10,15,100]
min_samples_leaf=[1,2,5,10]
```
